```python
#question 4 ,transpose of a image 

with open('face.png', 'rb') as f:
    header = f.read(54)
    data = bytearray(f.read())
width = int.from_bytes(header[18:22], 'little')
height = int.from_bytes(header[22:26], 'little')
transposed_data = bytearray(len(data))
for y in range(height):
    for x in range(width):
        i = (y * width + x) * 3
        j = (x * height + y) * 3
        transposed_data[j:j+3] = data[i:i+3]

# write the transposed image to a new file
with open('transposed_image.png', 'wb') as f:
    f.write(header)
    f.write(transposed_data)
```

```python
#QUESTION P2.1 ,LOAD,ENCODE,DECODE  JSON URL

from urllib.request import urlopen
import json
url = "https://jsonplaceholder.typicode.com/todos"
response = urlopen(url)
data = json.loads(response.read())
data
```

```python
encoded_data = json.dumps(data)
encoded_data
```

```python
decoded_data = json.loads(encoded_data)
decoded_data 
```

```python

# QUESTION P2.2 find the number of users
user_ids = set(todo['userId'] for todo in data)
num_users = len(user_ids)
print(f"Number of users: {num_users}")

```

```python
# QUESTION P2.3 find the number of tasks for each user
tasks_by_user = {user_id: 0 for user_id in user_ids}
for todo in data:
    user_id = todo['userId']
    tasks_by_user[user_id] += 1
for user_id, num_tasks in tasks_by_user.items():
    print(f"User {user_id}: {num_tasks} tasks")

```

```python
# QUESTION P2.4 find the number of completed and incomplete tasks for each user and rank them
completed_tasks_by_user = {user_id: 0 for user_id in user_ids}
incomplete_tasks_by_user = {user_id: 0 for user_id in user_ids}
for todo in data:
    user_id = todo['userId']
    if todo['completed']:
        completed_tasks_by_user[user_id] += 1
    else:
        incomplete_tasks_by_user[user_id] += 1
completed_tasks_by_user = dict(sorted(completed_tasks_by_user.items(), key=lambda x: x[1], reverse=True))
incomplete_tasks_by_user = dict(sorted(incomplete_tasks_by_user.items(), key=lambda x: x[1], reverse=True))
print("Ranking based on number of completed tasks:")
for rank, (user_id, num_completed) in enumerate(completed_tasks_by_user.items()):
    print(f"{rank+1}. User {user_id}: {num_completed} completed tasks")
print("Ranking based on number of incomplete tasks:")
for rank, (user_id, num_incomplete) in enumerate(incomplete_tasks_by_user.items()):
    print(f"{rank+1}. User {user_id}: {num_incomplete} incomplete tasks")
```

```python
# A.,B Connect to the camera and capture frames ,reconnects automatically when the camera goes down

import cv2
import time
import pyzbar.pyzbar as pyzbar


camera_ip = "http://0.0.0.0:8080/video"
cap = cv2.VideoCapture(camera_ip)
frame_count = 0
start_time = time.time()
fps = 0
while True:
    ret, frame = cap.read()
    if not ret:
        print("Camera went offline!")
        attempts = 0
        while True:
            cap = cv2.VideoCapture(camera_ip)
            ret, frame = cap.read()
            if ret:
                print(f"Camera is back online after {attempts} attempts.")
                break
            attempts += 1
            time.sleep(5)
    else:
        frame_count += 1
        if frame_count % 10 == 0:  
            end_time = time.time()
            elapsed_time = end_time - start_time
            fps = frame_count / elapsed_time
            latency = 1 / fps
            start_time = end_time
            frame_count = 0
        decoded_objects = pyzbar.decode(frame)
        for obj in decoded_objects:
            print(f"Found QR code with data: {obj.data.decode()}")
```

```python
# C. To trigger an alert message whenever the camera goes offline, you can use an event-based programming approach. For example, you can use the `threading` module to run the code above in a separate thread, and have the main thread periodically check if the camera is still sending frames. If it's not, you can trigger an alert message. Here's an example:

import threading

def capture_frames():
    while True:
        ret, frame = cap.read()
    if not ret:
        print("Camera went offline!")
        attempts = 0
        while True:
            cap = cv2.VideoCapture(camera_ip)
            ret, frame = cap.read()
            if ret:
                print(f"Camera is back online after {attempts} attempts.")
                break
            attempts += 1
            time.sleep(5)
    else:
        frame_count += 1
        if frame_count % 10 == 0:  
            end_time = time.time()
            elapsed_time = end_time - start_time
            fps = frame_count / elapsed_time
            latency = 1 / fps
            start_time = end_time
            frame_count = 0

# start capturing frames in a separate thread
t = threading.Thread(target=capture_frames)
t.start()

# periodically check if the camera is still sending frames
while True:
    if time.time() - start_time > 10:  # if it's been more than 10 seconds since the last frame
        print("ALERT: Camera is offline!")
    time.sleep(1)
```
